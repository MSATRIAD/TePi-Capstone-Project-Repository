```python
import pandas as pd

# Define a function to calculate the negative points based on the nutrition information
def calculate_negative_points(energy, sugars, saturated_fat, salt):
    # Energy (A)
    if energy <= 7.2:
        points_energy = 0
    elif energy > 7.2 and energy <= 14.3:
        points_energy = 1
    elif energy < 14.3:
        points_energy = 2
    elif energy < 21.5:
        points_energy = 3
    elif energy < 28.5:
        points_energy = 4
    elif energy < 35.9:
        points_energy = 5
    elif energy < 43.0:
        points_energy = 6
    elif energy < 50.2:
        points_energy = 7
    elif energy < 57.4:
        points_energy = 8
    elif energy < 64.5:
        points_energy = 9
    else:
        points_energy = 10

    # Sugars (B)
    if sugars <= 0:
        points_sugars = 0
    elif sugars < 1.5:
        points_sugars = 1
    elif sugars < 3.0:
        points_sugars = 2
    elif sugars < 4.5:
        points_sugars = 3
    elif sugars < 6.0:
        points_sugars = 4
    elif sugars < 7.5:
        points_sugars = 5
    elif sugars < 9.0:
        points_sugars = 6
    elif sugars < 10.5:
        points_sugars = 7
    elif sugars < 12.0:
        points_sugars = 8
    elif sugars < 13.5:
        points_sugars = 9
    else:
        points_sugars = 10

    # Saturated fat (C)
    if saturated_fat <= 1:
        points_saturated_fat = 0
    elif saturated_fat <= 2:
        points_saturated_fat = 1
    elif saturated_fat <= 3:
        points_saturated_fat = 2
    elif saturated_fat <= 4:
        points_saturated_fat = 3
    elif saturated_fat <= 5:
        points_saturated_fat = 4
    elif saturated_fat <= 6:
        points_saturated_fat = 5
    elif saturated_fat <= 7:
        points_saturated_fat = 6
    elif saturated_fat <= 8:
        points_saturated_fat = 7
    elif saturated_fat <= 9:
        points_saturated_fat = 8
    elif saturated_fat <= 10:
        points_saturated_fat = 9
    else:
        points_saturated_fat = 10


    # Salt (D)
    if salt < 0.09:
        points_salt = 0
    elif salt <= 0.18:
        points_salt = 1
    elif salt <= 0.27:
        points_salt = 2
    elif salt <= 0.36:
        points_salt = 3
    elif salt <= 0.45:
        points_salt = 4
    elif salt <= 0.54:
        points_salt = 5
    elif salt <= 0.63:
        points_salt = 6
    elif salt <= 0.72:
        points_salt = 7
    elif salt <= 0.81:
        points_salt = 8
    elif salt <= 0.90:
        points_salt = 9
    else:
        points_salt = 10

    return points_energy, points_sugars, points_saturated_fat, points_salt


# Define a function to calculate the positive points based on the nutrition information
def calculate_positive_points(fruits_vegetables, fiber, proteins):
    # Fruits, vegetables, nuts (E)
    if fruits_vegetables < 40:
        points_fruits = 0
    elif fruits_vegetables <= 60:  # For values between 40 and 60
        points_fruits = 2
    elif fruits_vegetables <= 80:  # For values between 60 and 80
        points_fruits = 4
    else:  # For values greater than 80
        points_fruits = 10

    # Fiber (F)
# Fiber (F)
    if fiber < 0.7:
        points_fiber = 0
    elif fiber <= 1.4:  # For values between 0.7 and 1.4
        points_fiber = 1
    elif fiber <= 2.1:  # For values between 1.4 and 2.1
        points_fiber = 2
    elif fiber <= 2.8:  # For values between 2.1 and 2.8
        points_fiber = 3
    elif fiber <= 3.5:  # For values between 2.8 and 3.5
        points_fiber = 4
    else:  # For values greater than 3.5
        points_fiber = 5


    # Proteins (P)
    if proteins < 1.6:
        points_proteins = 0
    elif proteins <= 3.2:  # For values between 1.6 and 3.2
        points_proteins = 1
    elif proteins <= 4.8:  # For values between 3.2 and 4.8
        points_proteins = 2
    elif proteins <= 6.4:  # For values between 4.8 and 6.4
        points_proteins = 3
    elif proteins <= 8.0:  # For values between 6.4 and 8.0
        points_proteins = 4
    else:  # For values greater than 8.0
        points_proteins = 5

    return points_fruits, points_fiber, points_proteins


# Define the function to calculate the NutriScore
def calculate_nutriscore(row):
    points_energy, points_sugars, points_saturated_fat, points_salt = calculate_negative_points(
        row['energy-kcal_100g'], row['sugars_100g'], row['saturated-fat_100g'], row['salt_100g'])
    
    points_fruits, points_fiber, points_proteins = calculate_positive_points(
        row['fruits-vegetables-nuts-estimate-from-ingredients_100g'], row['fiber_100g'], row['proteins_100g'])

    # Total negative and positive points
    negative_points = points_energy + points_sugars + points_saturated_fat + points_salt
    positive_points = points_fruits + points_fiber + points_proteins

    # Final score
    score = negative_points - positive_points

    # Determine NutriScore grade
    if score <= -1:
        grade = 'A'
    elif score <= 2:
        grade = 'B'
    elif score <= 10:
        grade = 'C'
    elif score <= 18:
        grade = 'D'
    else:
        grade = 'E'

    return points_energy, points_sugars, points_saturated_fat, points_salt, points_fruits, points_fiber, points_proteins, negative_points, positive_points, score, grade

import pandas as pd

# Load the real CSV data
file_path = "../ML/data/catalog.csv"
df = pd.read_csv(file_path)

# Apply the NutriScore calculation to each row
df[['energy_points', 'sugars_points', 'saturated_fat_points', 'salt_points',
    'fruits_points', 'fiber_points', 'proteins_points', 'total_negative_points',
    'total_positive_points', 'total_score', 'nutriscore_grade']] = df.apply(
    calculate_nutriscore, axis=1, result_type="expand"
)

# Save the result to a new CSV file
output_file_path = "../ML/data/catalog_final.csv"
df.to_csv(output_file_path, index=False)

print(f"Processed data saved to {output_file_path}")
```
